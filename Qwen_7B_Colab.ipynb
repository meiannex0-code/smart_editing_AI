```python
# =============================================================================
# STEP 2 COMPLETE: BILINGUAL CONTENT STUDIO CORE ENGINE - ENHANCED VERSION
# Qwen2.5-7B + User-Selected Simplified/Traditional Chinese
# =============================================================================

# Install required libraries
!pip install transformers torch opencc-python-reimplemented -q bitsandbytes accelerate

import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, AutoModelForCausalLM
import re
from opencc import OpenCC

print("📦 Libraries loaded!")

# Load Models
print("\n🤖 Loading MT for Translation...")
mt_model_name = "Helsinki-NLP/opus-mt-en-zh"  # English to Chinese MT
mt_tokenizer = AutoTokenizer.from_pretrained(mt_model_name)
mt_model = AutoModelForSeq2SeqLM.from_pretrained(mt_model_name)

print("\n🤖 Loading Qwen2.5-7B for Tone Editing...")
qwen_model_name = "Qwen/Qwen2.5-7B-Instruct"  # Upgraded to Qwen2.5-7B
qwen_tokenizer = AutoTokenizer.from_pretrained(qwen_model_name, trust_remote_code=True)
qwen_model = AutoModelForCausalLM.from_pretrained(
    qwen_model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True,
    load_in_4bit=True,  # 4-bit quantization for free Colab
    bnb_4bit_compute_dtype=torch.float16
)
print("✅ Models loaded!")

# OpenCC Converter for Simplified/Traditional
converter = None  # Will be set based on user choice

# Vibe Detection (Your original)
def detect_input_vibe(input_text):
    clean_text = re.sub(r'[^\w\s]', '', input_text.lower())
    vibe_keywords = {
        "melancholic": ["alone", "forgotten", "fading", "lost", "忧伤", "深夜"],
        "nostalgic": ["remember", "childhood", "old", "past", "怀念", "旧时光"],
        "urgent": ["now", "hurry", "limited", "立即", "抢购"],
        "playful": ["haha", "giggle", "silly", "fun", "哈哈", "逗趣"],
        "professional": ["expert", "proven", "专业", "权威"],
        "excited": ["amazing", "wow", "love", "超棒", "爱了"],
        "romantic": ["love", "heart", "亲爱的", "心动"],
        "chaotic": ["搞事", "整活", "笑死", "666"],
        "coquettish": ["宝贝", "哼", "人家", "撒娇"],
        "cringe": ["老铁", "双击", "土味", "亲"],
        "quirky": ["weird", "funny", "unexpected", "怪"],
        "heartfelt": ["story", "journey", "thank", "感动"]
    }
    for vibe, keywords in vibe_keywords.items():
        if any(kw in clean_text for kw in keywords):
            return vibe
    return "neutral"

# 16-Tone Dictionary (Your original)
tone_guidance = {
    "poetic": "Blend aesthetic and artsy: Use dreamy, metaphor-rich language with curated imagery, like indie film captions or Instagram moodboards. Evoke emotion through sensory details and rhythmic phrasing, drawing from ancient Chinese poetry traditions.",
    "storytelling": "Narrative-driven with nostalgic warmth: Structure as immersive beginning-middle-end journeys, blending memory-soaked longing with emotional arcs. Perfect for fanfic, screenplays, or heartfelt memoirs.",
    "dramatic": "Edgy with emo intensity: Use bold, over-exaggerated phrasing with moody, dark sentimentality. Create tension through rebellious attitude and anime-style monologues, ideal for cinematic scripts.",
    "playful": "Quirky with snappy wit: Mix childlike humor, wordplay, and sharp comebacks with meme-inspired irony. Keep it light-hearted for smile-inducing creative content.",
    "persuasive": "Urgent with social-proof: Use time-sensitive, benefit-focused language with community validation and testimonials. Build FOMO with calls-to-action for high-conversion ads.",
    "aspirational": "Authoritative with exclusive prestige: Paint visions of success and VIP status with confident, expert-led language. Inspire through elevated lifestyle goals.",
    "trust-building": "Friendly with heartfelt authenticity: Blend approachable peer recommendations with emotional, values-driven sincerity. Use reassuring phrasing to reduce doubt.",
    "eco-conscious": "Warm with sustainable vision: Combine gentle encouragement with ethical lifestyle aspirations. Highlight green impact for environmentally aware audiences.",
    "youth-hype": "Trendy with urgent energy: Mix Gen Z slang, emojis, and hype-driven phrasing with time-sensitive FOMO. Create viral content with 'yyds' vibes.",
    "casual": "Friendly with playful charm: Use relaxed, chatty language like trusted peer conversations, blended with quirky wordplay for WeChat or Douyin posts.",
    "bullet-comment": "弹幕风 with chaotic mischief: Short, fast, punchy remarks like Bilibili reactions, mixed with 整活-style trolling. Use emojis and '666' energy.",
    "coquettish": "撒娇 with flirty playfulness: Blend cute, coquettish whining with soft, charming humor for Douyin lifestyle content.",
    "kitschy": "土味 with meme humor: Use deliberately awkward, self-mocking phrasing with kitschy sincerity. Blend '老铁' vibes with viral meme references.",
    "minimalist": "Clean with sophisticated restraint: Use sparse, design-driven phrasing with calm elegance. Evoke wabi-sabi aesthetics for premium identities.",
    "quirky": "Offbeat with bold personality: Mix playful weirdness with edgy, ironic humor for memorable, unconventional brand voices.",
    "heartfelt": "Warm with authentic connection: Blend emotional, values-driven sincerity with gentle storytelling for genuine relationships.",
    "neutral": "Use natural, balanced language appropriate for the context without strong emotional coloring or cultural specificity."
}

# VIBE MODIFIERS (Your updated version)
vibe_modifier_rules = {
    "melancholic": {"poetic": "subtly", "storytelling": "deeply", "dramatic": "intensely"},
    "nostalgic": {"poetic": "harmoniously", "storytelling": "naturally", "dramatic": "bittersweetly"},
    "urgent": {"persuasive": "intensely", "youth-hype": "explosively", "exclusive": "compellingly"},
    "professional": {"aspirational": "confidently", "trust-building": "reliably"},
    "excited": {"youth-hype": "enthusiastically", "playful": "vibrantly"},
    "romantic": {"poetic": "passionately", "coquettish": "charmingly"},
    "chaotic": {"bullet-comment": "wildly", "playful": "uncontrollably"},
    "coquettish": {"coquettish": "charmingly", "playful": "flirtatiously"},
    "cringe": {"kitschy": "authentically", "casual": "self-consciously"},
    "quirky": {"quirky": "delightfully", "playful": "unexpectedly"},
    "heartfelt": {"heartfelt": "genuinely", "trust-building": "sincerely"},
    "rebellious": {"dramatic": "defiantly", "quirky": "boldly", "casual": "cheekily"},
    "sophisticated": {"minimalist": "elegantly", "aspirational": "refinedly", "eco-conscious": "thoughtfully"},
    "casual": {"casual": "naturally", "friendly": "warmly", "playful": "lightly"},
    "neutral": {t: "naturally" for t in tone_guidance}
}

# VIBE DESCRIPTIONS (Your updated version)
vibe_descriptions = {
    "melancholic": "emotional depth and subtle sadness",
    "nostalgic": "warm memories and reflective longing",
    "urgent": "time-sensitive energy and immediacy",
    "playful": "light-hearted fun and whimsy",
    "professional": "credible authority and precision",
    "excited": "vibrant enthusiasm and positive energy",
    "romantic": "intimate connection and heartfelt emotion",
    "rebellious": "bold attitude and countercultural edge",
    "casual": "relaxed familiarity and conversational flow",
    "sophisticated": "refined elegance and cultured sophistication",
    "quirky": "delightful weirdness and unexpected charm",
    "heartfelt": "genuine emotion and authentic connection",
    "chaotic": "energetic chaos and playful disruption",
    "coquettish": "charming flirtation and playful coyness",
    "cringe": "authentic awkwardness and self-aware humor"
}

# HYBRID TRANSLATION + TONE EDIT FUNCTION WITH USER CHOICE
def translate_and_tone_enhance(input_text, target_language="Chinese", tone="neutral", chinese_style="Simplified"):
    """Hybrid: MT for translation, Qwen for tone, user-selectable Simplified/Traditional."""
    global converter
    converter = OpenCC('t2s' if chinese_style == "Simplified" else 's2t')  # Set converter based on user choice

    # Language detection
    if re.search(r'[\u4e00-\u9fff]', input_text):
        source_language = "Chinese"
        if target_language == "Chinese":
            target_language = "English"
    else:
        source_language = "English"

    # Step 1: MT Translation
    if source_language == "English" and target_language == "Chinese":
        inputs = mt_tokenizer(input_text, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            translated_tokens = mt_model.generate(**inputs, max_length=200, num_beams=5, early_stopping=True)
        base_translation = mt_tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    elif source_language == "Chinese" and target_language == "English":
        inputs = mt_tokenizer(input_text, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            translated_tokens = mt_model.generate(**inputs, max_length=200, num_beams=5, early_stopping=True)
        base_translation = mt_tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    else:
        base_translation = input_text  # Same language case

    # Step 2: Vibe Detection
    detected_vibe = detect_input_vibe(input_text)

    # Step 3: Get base tone guidance
    base_guidance = tone_guidance.get(tone, tone_guidance["neutral"])

    # Step 4: Seamless tone enhancement with Qwen
    if detected_vibe != "neutral" and tone in vibe_modifier_rules.get(detected_vibe, {}):
        modifier = vibe_modifier_rules[detected_vibe][tone]
        vibe_desc = vibe_descriptions.get(detected_vibe, f"{detected_vibe} emotional resonance")
        enhanced_guidance = f"{base_guidance} {modifier} preserving the input's {vibe_desc}."
    else:
        enhanced_guidance = base_guidance

    # Step 5: Qwen tone edit prompt
    target_style = "Simplified Chinese (简体中文)" if chinese_style == "Simplified" else "Traditional Chinese (繁体中文)"
    if target_language == "Chinese":
        tone_prompt = f"""[INST] Edit this {target_style} text to apply the {tone} style:

Text: "{base_translation}"

Guidance: {enhanced_guidance}

Output only the tone-enhanced {target_style} translation: [/INST]"""
    else:
        tone_prompt = f"""[INST] Edit this {target_language} text to apply the {tone} style:

Text: "{base_translation}"

Guidance: {enhanced_guidance}

Output only the tone-enhanced {target_language} translation: [/INST]"""

    # Step 6: Generate toned output with Qwen
    tone_inputs = qwen_tokenizer(tone_prompt, return_tensors="pt", truncation=True, max_length=512)
    tone_inputs = {k: v.to(qwen_model.device) for k, v in tone_inputs.items()}

    with torch.no_grad():
        tone_outputs = qwen_model.generate(
            **tone_inputs,
            max_new_tokens=200,
            temperature=0.75,
            do_sample=True,
            top_p=0.9,
            top_k=50,
            repetition_penalty=1.1,
            pad_token_id=qwen_tokenizer.eos_token_id
        )

    # Extract and convert final translation
    final_translation = qwen_tokenizer.decode(tone_outputs[0], skip_special_tokens=True).strip()
    final_translation = re.sub(r'\n+', ' ', final_translation).strip()

    if target_language == "Chinese":
        final_translation = converter.convert(final_translation)  # Apply user-selected style

    return final_translation, detected_vibe

# QUICK TEST (Simulating Streamlit dropdown with "Simplified")
print("\n" + "="*60)
print("🧪 HYBRID VALIDATION TEST (Simplified Chinese)")
print("="*60)

test_input = "The old house creaked with forgotten laughter under the moonlight."
result, vibe = translate_and_tone_enhance(test_input, "Chinese", "poetic", "Simplified")

print(f"Input: {test_input}")
print(f"Detected Vibe: {vibe}")
print(f"Tone Applied: poetic")
print(f"Output: {result}")
print("\n✅ ENGINE READY!")

# Full test suite
print("\n" + "🔍 RUNNING COMPREHENSIVE TESTS..." + "\n")

test_cases = [
    {"input": "The old house creaked with forgotten laughter.", "tone": "poetic", "target": "Chinese", "style": "Simplified"},
    {"input": "Limited edition sneakers, only 50 pairs left!", "tone": "persuasive", "target": "Chinese", "style": "Simplified"},
    {"input": "这款奶茶超好喝姐妹们快来！", "tone": "bullet-comment", "target": "English", "style": "N/A"},
    {"input": "Our family farm grows coffee for 3 generations.", "tone": "heartfelt", "target": "Chinese", "style": "Traditional"},
    {"input": "New gaming mouse with RGB lighting!", "tone": "youth-hype", "target": "Chinese", "style": "Simplified"}
]

for i, case in enumerate(test_cases, 1):
    result, vibe = translate_and_tone_enhance(case["input"], case["target"], case["tone"], case.get("style", "Simplified"))
    print(f"\n📝 Test {i}: {case['tone']} → {case['target']} ({case.get('style', 'N/A')})")
    print(f"Vibe: {vibe} | Output: {result[:80]}...")
```
